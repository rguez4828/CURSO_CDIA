## 2. Análisis Exploratorio de datos

### Cargar Datos

```python
import pandas as pd

pd.set_option('future.no_silent_downcasting', True)

url_base = 'https://github.com/rguez4828/CURSO_CDIA/raw/refs/heads/master/Data/Coronario_Nacional/'
nombre_archivo = 'Base%20de%20datos%20coronario%204-9-21%20(6).xlsx'
ruta_archivo = url_base + nombre_archivo

coronario_df = pd.read_excel(ruta_archivo, index_col='Unnamed: 0')
```

```python
coronario_df.head()
```

```python
coronario_df.info()
```

Los datos fueron cargados correctamente con pandas, y segun la información del DataFrame hay **1370** filas con **62** columnas (variables). Del total de variables **47** son del tipo float64 y **15** del tipo object. La memoria total a utilizar será de más de **706.6 KB**

Al ser un dataset con datos medicos, en la base de datos existen variables que relacionan datos no públicos que no se van a tratar en el análisis preliminar de los datos, por lo que serán eliminados. Las columnas que se van a eliminar son:
* Nombre
* Los Apellidos
* HC, Numero de Historia Clinica
* Fecha de Ingreso y Egreso
* El tiempo de Estadía
* Descripción

Al mismo tiempo es necesario detectar filas y columnas duplicadas para eliminarlas

```python
coronario_df[coronario_df.duplicated(subset=['Nombre', '1er apell', '2do apell', 'HC', 'Sexo', 'Edad', 'Raza'])] # Deteccion de filas duplicadas por pacientes
```

```python
coronario_df[coronario_df.duplicated()]
```

Se detectaron un total de **73** filas con datos de pacientes duplicados de estas **21** filas con valores NaN. Solo se eliminarán las filas que tienen todos sus campos en NaN.

```python
coronario_df.dropna(how='all', inplace=True)  #Elimina filas que todos sus valores son NaN
```

```python
coronario_df.drop(['Nombre', '1er apell', '2do apell', 'HC', 'F ingreso', 'F egreso', 'Estadía', 'Descripcion '], axis=1, inplace=True) # Por columna especificada
```

Analizando el DataFrame existen campos con valores vacios (No NaN y No None) con un caracter "Espacio" y "-", que deben de ser sustituidos por None para que las variables puedan ser interpretadas correctamente.

Para esto se utilizará la expresión regular r'^\s*$' que remplazara cadenas vacías (''), cadenas con solo espacios (' ','  ' etc.) y cadenas con saltos de línea ('\n', '\n\n', etc.)

```python
coronario_df.replace(r'^\s*$', None, regex=True, inplace=True) # Sustitucion del caracter "Espacio"
coronario_df.replace('-', None, inplace=True)  # Sustitucion del caracteer "-"
```

Realizaremos el análisis univariado de algunas variables más importantes, para esto se creó la función **analisis_var_cualitativa()** que podrá ser reutilizada para cada variable cualitativa:

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def analisis_var_cualitativa(dataframe: pd.DataFrame):
    mean_df = dataframe.mean()
    mode_df = dataframe.mode()[0]
    median_df = dataframe.median()

    if mean_df > median_df and mean_df > mode_df:
        print(f"La distribución de la variable {dataframe.name} tiene asimetría positiva (cola hacia la derecha)")  # Media > Mediana > Moda
    elif mean_df < median_df and mean_df < mode_df:
        print(f"La distribución de la variable {dataframe.name} tiene asimetría negativa (cola hacia la izquierda)")  # Moda > Mediana > Media
    else:
        print(f"La distribución de la variable {dataframe.name} parece ser simétrica (Media ≈ Mediana ≈ Moda)")  # Asimetría cero

    _fig, _axs = plt.subplots(1, 2, figsize=(15, 3))

    # Calcular bins usando la regla de Freedman-Diaconis
    _q1 = dataframe.quantile(0.25)
    _q3 = dataframe.quantile(0.75)
    _iqr = _q3 - _q1
    _bin_width = 2 * _iqr / (len(dataframe) ** (1/3))
    _bins = int((dataframe.max() - dataframe.min()) / _bin_width)

    # Histograma con curva de distribución normal
    _axs[0].set_title("Histograma para " + dataframe.name)
    dataframe.plot(kind='hist', bins=_bins, ax=_axs[0], density=True, alpha=0.6) # Hace que el histograma muestre densidades en lugar de frecuencias absolutas, lo cual permite compararlo con la curva de densidad

    # Calcular parámetros de la distribución normal
    _mu, _std = dataframe.mean(), dataframe.std()

    # Generar puntos x para la curva de densidad
    _x = np.linspace(dataframe.min(), dataframe.max(), 100)
    _p = stats.norm.pdf(_x, _mu, _std) # Calcula la función de densidad de probabilidad (PDF) de una distribución normal con media mu y desviación estándar std

    # Añadir la curva de densidad
    _axs[0].plot(_x, _p, 'k', linewidth=2)  # 'k' es color negro

    # Etiquetas y leyenda
    _axs[0].legend(['Datos Observados', 'Distribución Normal'])

    # Box Plot
    _axs[1].set_title("Box Plot para " + dataframe.name)
    dataframe.plot(kind='box', vert=False, ax=_axs[1])

    plt.tight_layout()
```

```python
x_Creat = coronario_sel_csv_df[['Creat']]
y_Ac_U = coronario_sel_csv_df[['Ac U']]

def scatter_hist(x, y, ax, ax_histx, ax_histy):
    # no labels
    ax_histx.tick_params(axis="x", labelbottom=False)
    ax_histy.tick_params(axis="y", labelleft=False)

    x_label = x.columns[0]  # Tomamos el nombre de la columna como etiqueta
    y_label = y.columns[0]

    # Determinar límites comunes para x e y
    x_min, x_max = x.min().values[0], x.max().values[0]
    y_min, y_max = y.min().values[0], y.max().values[0]

    common_min = min(x_min, y_min) - 10
    common_max = max(x_max, y_max) + 10

    # Gráfico de dispersión
    ax.scatter(x, y, c='skyblue', alpha=0.6)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

    # Establecer límites iguales para x e y en el scatter
    ax.set_xlim(common_min, common_max)
    ax.set_ylim(common_min, common_max)

    # now determine nice limits by hand:
    binwidth = 0.25
    xymax = max(np.max(np.abs(x)), np.max(np.abs(y)))
    lim = (int(xymax/binwidth) + 1) * binwidth

    bins = np.arange(-lim, lim + binwidth, binwidth)
    ax_histx.hist(x, bins=bins, edgecolor='salmon', color='lightgray', alpha=0.6)
    ax_histx.set_xlim(common_min, common_max)  # Ajustar límite del histograma X
    ax_histy.hist(y, bins=bins, orientation='horizontal', edgecolor='salmon', color='lightgray', alpha=0.6)
    ax_histy.set_ylim(common_min, common_max)  # Ajustar límite del histograma Y

fig, axs = plt.subplot_mosaic([['histx', '.'],
                               ['scatter', 'histy']],
                              figsize=(6, 6),
                              width_ratios=(4, 1), height_ratios=(1, 4),
                              layout='constrained')

scatter_hist(x_Creat, y_Ac_U, axs['scatter'], axs['histx'], axs['histy'])
```

Se contabilizaron cuantos valores faltan por cada variable, reutilizando código se emplea el método siguiente que ayuda a entender el fenómeno de los datos faltantes en el DataFrame.

```python
def tabla_valores_faltantes(df):
  """
    Genera una tabla con el número y porcentaje de valores faltantes por columna en un DataFrame.

    Parámetros:
    df (pd.DataFrame): DataFrame de entrada.

    Retorna:
    pd.DataFrame: Tabla con columnas que tienen valores faltantes, indicando el número y porcentaje de valores faltantes.
  """
  faltantes_suma = df.isnull().sum() # total de valores faltantes
  faltantes_porc = 100 * (faltantes_suma / len(df)) # porcentaje de valores faltantes
  faltantes_df = pd.concat([faltantes_suma, faltantes_porc], axis=1) # tabla de resultados
  faltantes_df = faltantes_df.rename( # renombramos columnas
        columns = {0 : 'Valores faltantes', 1 : '% del total de valores'}, )

  faltantes_df = faltantes_df[ # ordenar
            faltantes_df.iloc[:,1] != 0].sort_values(
        '% del total de valores', ascending=False).round(1)
  print ("El Dataframe tiene " + str(df.shape[1]) + " columnas.\n"
            "Hay " + str(faltantes_df.shape[0]) +
              " Columnas que tienen valores faltantes.")

  return faltantes_df
```
